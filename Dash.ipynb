```python
import plotly.express as px
import plotly.graph_objects as go
```

```python
import dash
from dash import dcc
from dash import html
from dash.dependencies import Input,Output
```

```python
df = px.data.stocks()
df
```

```python
app = dash.Dash(__name__)
app.config.suppress_callback_exceptions = True
```

```python
## Diseño de la aplicacion
app.layout = html.Div([
    html.Div([
        html.H1('Prueba de RECA'),
        html.Img(src='./imgs_plotly/logo_plotly.svg')
    ],className = 'banner'),
    
    html.Div([
        html.Div([
            html.P('Selecciona la marca',className='fix_label',style={'color':'black','margin_top':'2px'}),
            dcc.RadioItems(id='empresa',
                          labelStyle = {'display':'inline-block'},
                          options=[
                              {'label':'Google', 'value': 'GOOG'},
                              {'label':'Apple', 'value': 'AAPL'},
                              {'label':'Amazon', 'value': 'AMZN'},
                              {'label':'Facebook', 'value': 'FB'},
                              {'label':'Netflix', 'value': 'NFLX' }
                          ],value='GOOG', style={'tex-align':'center','color':'black'}, className = 'dcc_compon')
        ], className='create_container2 five columns',style={'margin-bottom':'20px'})
    ], className='row flex-display'),
    
    html.Div([
        html.Div([
            dcc.Graph(id='bar_graph',figure={})
        ], className='create_container2 eight columns'),
        
        html.Div([
            dcc.Graph(id='pie_graph',figure={})
        ], className='create_container2 five columns')
    ], className='row flex-display')
],id='main_container', style={'display':'flex','flex-direction':'column'})
```

```python
## Puente entre grafico y contenedor
@app.callback(
    Output('bar_graph',component_property='figure'),
    [Input('empresa',component_property='value')])

def update_graph_bar(value):

    if value == 'GOOG':
        fig = px.bar(
            data_frame = df,
            x = 'date',
            y = 'GOOG'
        )
    elif value == 'AAPL':
        fig = px.bar(
            data_frame = df,
            x = 'date',
            y = 'AAPL'
        )
    elif value == 'AMZN':
        fig = px.bar(
            data_frame = df,
            x = 'date',
            y = 'AMZN'
        )
    elif value == 'FB':
        fig = px.bar(
            data_frame = df,
            x = 'date',
            y = 'FB'
        )
    else:
        fig = px.bar(
            data_frame = df,
            x = 'date',
            y = 'NFLX'
        )
    
    return fig
```

```python
if __name__ == '__main__':
    app.run_server() # Para que se ejecute en el navegador
```
