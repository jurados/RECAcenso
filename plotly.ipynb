```python
import plotly.express as px
```

## Linear plot

```python
df = px.data.stocks()
df
```

```python
#Funciona sin el fig = px.line, solo con px.line ....
fig = px.line(df, x='date',y='NFLX')#, labels={'x':'Date','y':'Price'})
fig
```

```python
fig = px.line(df,x='date',y=['NFLX','AMZN'], labels={'x':'Date','y':'Price'},title='Netflix vs Amazon')
fig
```

## Scatter plot

```python
df = px.data.iris()
#df
```

```python
fig = px.scatter(df, x="sepal_width", y="sepal_length")#,color="species",size="petal_length",
                #hover_data=['petal_width']) # hover_data informacion extra
fig
```

```python
fig = px.scatter_3d(df, x='sepal_length',y='petal_length',z='petal_width',color='species',size='sepal_width')
fig
```

## ComparaciÃ³n multiple

```python
fig = px.scatter_matrix(df,dimensions=["sepal_width",'sepal_length','petal_width','petal_length'],color='species')
fig
```

## Bloxplot

```python
fig = px.box(df,x='species',y='sepal_width',color='species') # Se puede usar solo el valor de y sin necesidad del de x
fig
```

## Histograms

```python
fig = px.histogram(df, x='sepal_width',color='species')
fig.update_layout(barmode='overlay')
fig.update_traces(opacity=0.75)
fig
```

## Histogram: Counting

```python
fig = px.histogram(df,x='species',y='sepal_width',histfunc='count',height=300,title='Histrogram Chart')
fig
```

## Bar plot

```python
fig = px.bar(df,x='species')
fig
```

```python
fig = px.bar(df,x='species',y='sepal_length')
fig
```

## Scatter plot with regression liens

```python
fig = px.scatter(df,x='sepal_width',y='sepal_length',color='species',trendline="ols")#,template='simple_white')
# es necesario tener instalado statsmodels
fig
```

## Scatterplot adding marginal plots

```python
fig = px.scatter(df,x='sepal_width',y='sepal_length',color='species',
                 marginal_x='box',marginal_y='box',trendline='ols',template='ggplot2')#,template='simple_white')
# es necesario tener instalado statsmodels
fig
```
